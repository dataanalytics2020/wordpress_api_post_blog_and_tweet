```python
# -*- encoding:utf-8 -*-

from sqlalchemy import (Column, String, Text, ForeignKey, \
                create_engine, MetaData, DECIMAL, DATETIME, exc, event, Index)
from sqlalchemy.schema import UniqueConstraint
from sqlalchemy.orm import (sessionmaker, relationship, scoped_session)
from sqlalchemy.ext.declarative import declarative_base
from sqlalchemy.dialects.mysql import INTEGER as Integer
from datetime import datetime
import os

username = os.getenv('DB_USER_NAME')
password=os.getenv('DB_PASSWORD')
host=os.getenv('DB_HOST')
port='3306'
database=os.getenv('DB_NAME')
db_url = f"mysql+mysqlconnector://{username}:{password}@{host}:{port}/{database}"

# DB接続するためのEngineインスタンス
engine = create_engine(db_url, echo=False)

Session = scoped_session(sessionmaker(autocommit=False, autoflush=False, bind=engine))

metadata = MetaData()
Base = declarative_base()
```

```python
from sqlalchemy import Column, Integer, create_engine, MetaData, DATETIME, String,Date
from sqlalchemy.ext.declarative import declarative_base
from sqlalchemy.orm import sessionmaker
from datetime import datetime
from pathlib import Path
import os
import datetime 
import pandas as pd
```

```python
class Schedule(Base):
    __tablename__ = "schedule"  # テーブル名を指定
    __table_args__ = {"mysql_engine": "InnoDB"}

    schedule_id = Column('id',Integer, primary_key=True, autoincrement=True)
    prefecture = Column('都道府県',String(255))
    coverage_date = Column('イベント日',Date, nullable=False)
    day_of_week = Column('曜日',String(1), nullable=False)
    parlor_name = Column('店舗名',String(255), nullable=False)
    coverage_name = Column('取材名',String(255), nullable=False)
    medium_name = Column('媒体名',String(255))
    schedule_rank = Column('取材ランク',String(1))
    created = Column("取得時間", DATETIME, default=datetime.datetime.now, nullable=False)
    def __init__(self):
        now = datetime.datetime.now()
        self.created = now

    def record_detail(self):  # フルネームを返すメソッド
        return f"{self.coverage_date} {self.parlor_name} {self.coverage_name} "

Base.metadata.create_all(bind=engine)
```

```python
class Pledge(Base):
    __tablename__ = "pledge"  # テーブル名を指定
    __table_args__ = {"mysql_engine": "InnoDB"}

    pledge_id = Column('id',Integer, primary_key=True, autoincrement=True)
    coverage_name = Column('取材名',String(255), nullable=False)
    medium_name = Column('媒体名',String(255))
    pledge_content = Column('公約内容',String(255), nullable=True)
    created = Column("取得時間", DATETIME, default=datetime.datetime.now, nullable=False)
    def __init__(self):
        now = datetime.datetime.now()
        self.created = now

    def record_detail(self):  # フルネームを返すメソッド
        return f"{self.coverage_name} {self.coverage_name} "

Base.metadata.create_all(bind=engine)
```

```python
pledge_df = pd.read_csv(r'取材公約用マスター.csv')
pledge_df = pledge_df.fillna('')
pledge_df
```

```python
for i,row in pledge_df.iterrows():
    pledge = Pledge()
    print(i,row)
    #break
    pledge.coverage_name = row['取材名']
    pledge.medium_name = row['媒体名']
    pledge.pledge_content = row['公約内容']
    session.add(pledge)
    session.commit()
```
